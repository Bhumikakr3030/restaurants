```python
import pandas as pd

df = pd.read_csv("/content/final_food_delivery_dataset.csv")

df.head()
```

```python
gold_city_revenue = (
    df[df["membership"] == "Gold"]
    .groupby("city")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

gold_city_revenue
```

```python
cuisine_avg_order = (
    df.groupby("cuisine")["total_amount"]
    .mean()
    .sort_values(ascending=False)
)

cuisine_avg_order
```

```python
user_total_spend = (
    df.groupby("user_id")["total_amount"]
    .sum()
)

count_users = (user_total_spend > 1000).sum()
count_users
```

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ["3.0–3.5", "3.6–4.0", "4.1–4.5", "4.6–5.0"]

df["rating_range"] = pd.cut(df["rating"], bins=bins, labels=labels, include_lowest=True)

rating_revenue = (
    df.groupby("rating_range")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

rating_revenue
```

```python
gold_city_avg = (
    df[df["membership"] == "Gold"]
    .groupby("city")["total_amount"]
    .mean()
    .sort_values(ascending=False)
)

gold_city_avg
```

```python
cuisine_stats = (
    df.groupby("cuisine")
    .agg(
        restaurant_count=("restaurant_id", "nunique"),
        revenue=("total_amount", "sum")
    )
    .sort_values("restaurant_count")
)

cuisine_stats
```

```python
total_orders = len(df)
gold_orders = len(df[df["membership"] == "Gold"])

percentage = round((gold_orders / total_orders) * 100)
percentage
```

```python
restaurant_stats = (
    df.groupby("restaurant_name_y")
    .agg(
        total_orders=("total_amount", "count"),
        avg_order_value=("total_amount", "mean")
    )
)
```

```python
filtered = restaurant_stats[
    restaurant_stats["total_orders"] < 20
].sort_values("avg_order_value", ascending=False)

filtered.head(10)
```

```python
restaurant_map = (
    df[["restaurant_name_y", "restaurant_name_x"]]
    .drop_duplicates()
    .set_index("restaurant_name_y")["restaurant_name_x"]
)
```

```python
filtered["restaurant_name_x"] = filtered.index.map(restaurant_map)
```

```python
final_answer = filtered.reset_index()
final_answer
```

```python
df.groupby(["restaurant_name_y", "restaurant_name_x"])["total_amount"] \
  .agg(total_orders="count", avg_order_value="mean") \
  .query("total_orders < 20") \
  .sort_values("avg_order_value", ascending=False)
```

```python
combo_revenue = (
    df.groupby(["membership", "cuisine"])["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

combo_revenue
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["quarter"] = df["order_date"].dt.to_period("Q")

quarter_revenue = (
    df.groupby("quarter")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

quarter_revenue
```

```python
# Revenue per city for Gold members
gold_city_revenue = df[df['membership'] == 'Gold'].groupby('city')['total_amount'].sum()
top_gold_city = gold_city_revenue.idxmax()

# Orders in that city
orders_top_gold_city = df[(df['membership'] == 'Gold') & (df['city'] == top_gold_city)].shape[0]
orders_top_gold_city
```
